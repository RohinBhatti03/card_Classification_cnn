```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Conv2D,MaxPool2D,Dropout,Flatten,BatchNormalization
from keras import Sequential
from keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    
)
```

```python
val_datagen = ImageDataGenerator(
    rescale=1./255,
    
)
```

```python
# image_preprocess=ImageDataGenerator(
#     rescale=1./255
# )
```

```python
train_generator=train_datagen.flow_from_directory(
    "train",
    target_size=(128,128),
    batch_size=32,
    class_mode="categorical"
)
```

```python
val_generator=val_datagen.flow_from_directory(
    "valid",
    target_size=(128,128),
    batch_size=32,
    class_mode="categorical"
)
```

```python
#                                                      cnn model building 
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),activation="relu",padding="same",input_shape=(128,128,3)))
# model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(64,kernel_size=(3,3),activation="relu",padding="same"))
# model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(128,kernel_size=(3,3),activation="relu",padding="same"))
# model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2)))
#                                                  aff flatten layer for 1d array like size
model.add(Flatten())
#                                                    ann model building
model.add(Dense(128,activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(64,activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(53 ,activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer="Adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
early_stop=EarlyStopping(
    patience=5,
    monitor="val_loss",
    restore_best_weights=True,
    min_delta=0.001,
    mode="min"
)
```

```python
model.fit(train_generator , validation_data=val_generator,epochs=150,callbacks=[early_stop])
```
